```python
import os
import sys
import requests as req
import subprocess
import json
import pandas as pd

import seaborn as sn
from matplotlib import pyplot as plt
```

```python
out = subprocess.run(["bash", "curl.sh"], stdout=subprocess.PIPE).stdout
out = json.loads(out)
out['data']['lineup']['positions']
```

```python
arrayDef = []
arrayMed = []
arrayDel = []

arrayDefP = []
arrayMedP = []
arrayDelP = []

cont = 0

for i in range(1,5):
    if i == 1:
        arrayPort  = [out['data']['lineup']['positions']["1"]["1"]['name']]
        arrayPortP = [out['data']['lineup']['positions']["1"]["1"]['points']]
    elif i == 2:
        for j in range(len(out['data']['lineup']['positions']["2"])):
            #print(out['data']['lineup']['positions']["2"][str(j+2)]['name'])
            arrayDef  = arrayDef  + [out['data']['lineup']['positions']["2"][str(j+2)]['name']]
            arrayDefP = arrayDefP + [out['data']['lineup']['positions']["2"][str(j+2)]['points']]
    elif i == 3:
        for j in range(len(out['data']['lineup']['positions']["3"])):
            #print(out['data']['lineup']['positions']["3"][str(j+5)]['name'])
            arrayMed  = arrayMed  + [out['data']['lineup']['positions']["3"][str(j+5)]['name']]
            arrayMedP = arrayMedP + [out['data']['lineup']['positions']["3"][str(j+5)]['points']]
    elif i == 4:
        for j in range(len(out['data']['lineup']['positions']["4"])):
            #print(out['data']['lineup']['positions']["4"][str(j+10)]['name'])
            arrayDel  = arrayDel  + [out['data']['lineup']['positions']["4"][str(j+10)]['name']]
            arrayDelP = arrayDelP + [out['data']['lineup']['positions']["4"][str(j+10)]['points']]
    #guardas la j
```

```python
alineacion = pd.DataFrame(columns = ["Jugador", "Puntos"])

#alineacion.Defensa = arrayDef
alineacion.Jugador = arrayPort  + arrayDef  + arrayMed  + arrayDel
alineacion.Puntos  = arrayPortP + arrayDefP + arrayMedP + arrayDelP

alineacion
```
