```python
goles                = [11,11,7,7,3,1,5,3,7,1]
jugador_azul         = [10,8 ,5,6,6,1,1,3,3,1]
participante         = ["Juan Carlos", "Fran", "AndresPB8", "juamonchi", "Roberto", "Cortijo", "RulilloElPillo", "Adri4K_Mister", "_.jorgeortiz._", "pablito"]
expulsiones          = [2,0,0,0,0,0,1,0,0,0]
puntuacion_negativa  = [18,0,2,7,2,3,7,5,2,0]
jugadores_negativos  = [18,0,2,3,2,3,7,1,2,0]
n_jornadas           = [6,6,6,6,6,6,6,6,6,3]
jugadores_grises     = [8,4,9,7,4,9,10,10,12,8]
puntos               = [297,305,248,245,233,221,202,205,183,89]

df = pd.DataFrame(columns=["Posicion", "Participante", "Puntos", "Número_de_goles", "Número_de_Jugadores_en_Azul", "Expulsiones", "Número_de_Jornadas", "Jugadores_Grises",
                           "Puntuacion_Negativa", "Jugadores_Negativos"])

df.Participante = participante
df.Puntos = puntos
df.Número_de_goles = goles
df.Número_de_Jugadores_en_Azul = jugador_azul
df.Expulsiones = expulsiones
df.Número_de_Jornadas = n_jornadas
df.Jugadores_Grises = jugadores_grises
df.Puntuacion_Negativa = puntuacion_negativa
df.Jugadores_Negativos = jugadores_negativos
df.index = participante

df = df.sort_values(by=['Puntos'], ascending=False)

df.Posicion = ["Primero", "Segundo", "Tercero", "Cuarto", "Quinto", "Sexto", "Septimo", "Octavo", "Noveno", "Decimo"]
df
```
